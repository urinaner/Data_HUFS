##한국환경공단_에어코리아_측정소정보
https://www.data.go.kr/iim/api/selectAPIAcountView.do


```python
import requests
from urllib.parse import urlencode
from xml.etree import ElementTree
import pandas as pd

url = 'http://apis.data.go.kr/B552584/MsrstnInfoInqireSvc/getMsrstnList'
params ={'serviceKey' : 'MqKdo6Iq6kdusnVpSyqabRr7+gT0+RaI4UdmAWAMTTDuYx3P+AAnbZLek7v2DgcFmgOwHeirdrwtxYeTYQOkwA==', 'returnType' : 'xml', 'numOfRows' : '100', 'pageNo' : '1', 'addr' : '경기도', 'stationName' : '일산서구' }

# 인코딩된 파라미터를 문자열로 변환하여 출력
encoded_params = urlencode(params)
print(url + "?" + encoded_params)

response = requests.get(url, params=params)
#print(response.content)

root=ElementTree.fromstring(response.content)

df=pd.DataFrame()
for item in root.iter('item'):
  item_dict={}
  item_dict['dmX']=item.find('dmX').text
  item_dict['item']=item.find('item').text
  item_dict['mangName']=item.find('mangName').text
  item_dict['year']=item.find('year').text
  item_dict['addr']=item.find('addr').text
  item_dict['stationName']=item.find('stationName').text
  item_dict['dmY']=item.find('dmY').text
  df=df.append(item_dict, ignore_index=True)

df
```

```python
import requests
from urllib.parse import urlencode
from xml.etree import ElementTree
import pandas as pd

skey = 'MqKdo6Iq6kdusnVpSyqabRr7+gT0+RaI4UdmAWAMTTDuYx3P+AAnbZLek7v2DgcFmgOwHeirdrwtxYeTYQOkwA=='

url = 'http://apis.data.go.kr/B552584/ArpltnInforInqireSvc/getMinuDustFrcstDspth'
params = {
    'servicekey': skey,
    'returnType': 'xml',
    'numOfRows': '100',
    'pageNo': '1',
    'searchDate': '2023-07-24',
    'InformCode': 'PM10'
}

# 인코딩된 파라미터를 문자열로 변환하여 출력
encoded_params = urlencode(params)
print(url + "?" + encoded_params)

response = requests.get(url, params=params)
#print(response.content)

root=ElementTree.fromstring(response.content)

df=pd.DataFrame()
for item in root.iter('item'):
  item_dict={}
  item_dict['informCause']=item.find('informCause').text
  item_dict['informOverall']=item.find('informOverall').text
  item_dict['informData']=item.find('informData').text
  item_dict['informGrade']=item.find('informGrade').text
  item_dict['dataTime']=item.find('dataTime').text
  df=df.append(item_dict, ignore_index=True)

df
```
