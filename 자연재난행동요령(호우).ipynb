```python
from bs4 import BeautifulSoup
import urllib.request
import lxml
import requests
import re
```

```python
url="https://www.safekorea.go.kr/idsiSFK/neo/sfk/cs/contents/prevent/prevent18.html?menuSeq=126"

response = requests.get(url)
response.encoding = 'utf-8'  # Set the response encoding to UTF-8
html = response.text

soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
details=soup.find_all(['p'])
for detail in details:
  print(detail.get_text())
```

```python
contents=soup.find('div',{'class': 'detail'})
print(contents.get_text())
```

```python
contents=soup.find('div',{'class': 'tabmenu01'})
print(contents.get_text())
```

```python
contents=soup.find('div',{'class': 'tabarea'})
print(contents.get_text())
```

```python
context=soup.find('div', {'class':'section01'})
print(context.get_text())
```
